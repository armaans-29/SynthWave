```python
X = data.loc[:, ['Age', 'Smoking_Status', 'Employment_Status', 'Number Of Dependants']]
y = data['Annual_Premium_Amount']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
X.head()
```

```python
X['Employment_Status'] = le.fit_transform(X['Employment_Status'])
X['Smoking_Status'] = le.fit_transform(X['Smoking_Status'])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.predict([[26,	1,	1,	0]])
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
data_new = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
data_new
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

# Question 7

```python
data_num = ["Age", "Number Of Dependants", "Income_Lakhs", "Annual_Premium_Amount"]

co_rel = data[data_num].corr(method="pearson")
print("Correlation matrix:\n", co_rel)

plt.figure(figsize=(6, 4))
sns.heatmap(co_rel, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap", fontstyle = 'italic', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
premium_co_rel = co_rel["Annual_Premium_Amount"].drop("Annual_Premium_Amount")
premium_co_rel
```

```python
strongest_pos = premium_co_rel.idxmax()
strongest_neg = premium_co_rel.idxmin()

print("Strongest positive correlation with premium:", strongest_pos, premium_co_rel[strongest_pos])
print("Strongest negative correlation with premium:", strongest_neg, premium_co_rel[strongest_neg])
```

# Question 8

```python
cutoff = data["Annual_Premium_Amount"].quantile(0.75)
data["is_high_cost"] = data["Annual_Premium_Amount"] >= cutoff
data["is_high_cost"].sample(5)
```

```python
region_high = (
      data.groupby("Region")["is_high_cost"]
      .mean()
      .reset_index()
      .rename(columns={"is_high_cost": "high_cost_share"})
)

print("High-cost shared by Region:")
print(region_high)
```

```python
overall_share = data["is_high_cost"].mean()
print("Overall high-cost share:", overall_share)

above_avg_region = region_high[region_high["high_cost_share"] > overall_share]
print("Regions which are above average high-cost concentration : ")
print(above_avg_region)
```

```python
age_smoker = data.groupby(['Age', 'Smoking_Status'])
```

```python
sii = smoking/non_smoking
```

```python
sii
```

```python
sii.plot(kind='bar')
plt.title('Smoking Impact Index vs Age')
plt.xlabel('Age')
plt.ylabel('Smoking Impact Index')
plt.show()
```
